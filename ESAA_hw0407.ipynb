## **시간 범위 수정하고 데이터 밀어내기 - shift 메서드**
**1) 에볼라의 확산 속도 비교하기**

```python
#1 - X축을 Date 열로,y축을 사망자 수로 지정하여 그래프 그리기
ebola=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/esaa/자료/country_timeseries.csv',parse_dates=[0])

import matplotlib.pyplot as plt

ebola.index = ebola['Date']

fig,ax=plt.subplots()
ax = ebola.iloc[0:,1:].plot(ax=ax)
ax.legend(fontsize=7, loc=2, borderaxespad=0.)
plt.show()
```

```python
#2 - 각 나라의 에볼라 발병일이 달라 그래프가 그려지기 시작한 지점이 다르다.
# => 각 나라의 발병일을 가장 처음 에볼라가 발병한 Guinea와 동일하 위치로 옮겨야 나라별 에볼라 확산속도를 제대로 비교 가능
ebola_sub = ebola[['Day', 'Cases_Guinea','Cases_Liberia']]
print(ebola_sub.tail(10))
```

```python
#3 - 그래프를 그리기 위한 데이터프레임 준비하기 
# 중간에 아예 날짜가 업슨 데이터를 포함시켜 확산 속도를 제대로 비교 
ebola=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/esaa/자료/country_timeseries.csv',parse_dates=[0])
print(ebola.head().iloc[:,:5])
print(ebola.tail( ).iloc[:, :5])
```

```python
#4 - Date 열의 최댓값, 최솟값으로 시간 범위를 생성해 new_idx에 저장하기 
ebola.index=ebola['Date']
new_idx=pd.date_range(ebola.index.min(), ebola.index.max())
```

```python
#5 - new_idx를 살펴보면 ebola 데이터 집합에 있는 시간 순서와 반대로 생성되어 있음 => reverse()
print(new_idx)
new_idx = reversed(new_idx)
```

```python
#6 - reindex()를 사용하여 새로 생성한 인덱스를 새로운 인덱스로 지정
ebola=ebola.reindex(new_idx)
print(ebola.head().iloc[:,:5])
print(ebola.tail().iloc[:,:5])
```

```python
#7 - 각 나라의 에볼라 발병일 옮기기
last_valid = ebola.apply(pd.Series.last_valid_index) 
print(last_valid)
```

```python
#8 - 각 나라의 에볼라 발병일을 동일한 출발선으로 옮기려면 에볼라가 가장 처음 발병한 날에서 각 나라의 에볼라 발병일을 뺀 만큼 옮긴다.
earliest_date = ebola.index.min( ) 
print(earliest_date)

shift_values=last_valid-earliest_date
print(shift_values)
```

```python
#9 - 각 나라의 에볼라 발병일 옮기기 
ebola_dict = {}
for idx, col in enumerate(ebola):
  d = shift_values[idx].days
  shifted = ebola[col].shift(d) 
  ebola_dict[col] = shifted
```

```python
#10 - ebola_dict에는 시간을 다시 설정한 데이터가 딕셔너리 형태로 저장되어 있음
# ebola_dict의 값을 데이터프레임으로 변환하기
ebola_shift = pd.DataFrame(ebola_dict)
```

```python
#11 - 에볼라 최초 발병일을 기준으로 모든 열의 데이터가 옮겨짐을 확인 
print(ebola_shift.tail())
```

```python
#12 - 마지막 인덱스를 Day 열로 지정하고 필요없는 열은 삭제
ebola_shift.index = ebola_shift['Day']
ebola_shift = ebola_shift.drop(['Date', 'Day'],axis=1)
print(ebola_shift.tail( ))
```

```python
#13- 지금까지 만든 데이터프레임으로 다시 그래프 그리기
fig, ax = plt.subplots( ) 
ax = ebola_shift.iloc[:, :].plot(ax=ax) 
ax.legend(fontsize=7, loc=2, borderaxespad=0.)
plt.show( )
```
