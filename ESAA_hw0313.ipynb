## **DataFrame 데이터 삭제**
drop() 메서드를 이용해 삭제한다. 
* DataFrame.drop(labels=None, axis=0, index=None, columns=None, level=None, inplace=False, errors='raise')
* axis는 특정 칼럼 또는 특정 행을 드롭하며, axis 0은 로우방향축, axis 1은 칼럼 방향 축이다.

```python
# 'Age_0' 칼럼 삭제하기 
titanic_drop_df=titanic_df.drop('Age_0', axis=1)
titanic_drop_df.head(3)
```

```python
titanic_df.head(3)
```

* 'Age_0' 칼럼이 여전히 존재함을 볼 수 있음. 
* 그 이유는 inplace=False로 설정했기 때문이다. (디폴트가 False) inplace=True로 설정하면 자신의 데이터프레임에서 데이터를 삭제한다. 
* 여러개의 칼럼을 삭제하고 싶다면 리스트 형태로 삭제하고자 하는 칼럼명을 입력해 labels 파라미터로 입력한다. 

```python
drop_result=titanic_df.drop(['Age_0',	'Age_by_10',	'Family_No'], axis=1, inplace=True)
print('inplace=True로 drop한 후 반환된 값: ', drop_result)
titanic_df.head(3)
```

```python
pd.set_option('display.width', 1000)
pd.set_option('display.max_colwidth', 15)
print('#### before axis 0 drop ####')
print(titanic_df.head(3))
titanic_df.drop([0,1,2], axis=0, inplace=True)
print('#### after axis0 drop ####')
print(titanic_df.head(3))
```

**<정리>**
* axis: 로우를 삭제할 때는 axis=0, 칼럼을 삭제할 때는 axis=1으로 설정
* 원본 데이터프레임은 유지하고 드롭된 데이터프레임을 새롭게 객체 변수로 받고 싶다면 inplace=False로 설정
* 원본 데이터프레임에 드롭된 결과를 적용할 경우에는 inplace=True를 적용
* 원본 데이터프레임에서 드롭된 데이터프레임을 다시 원본 데이터프레임 객체 변수로 할당하면 원본 데이터프레임에서 드롭된 결과를 적용할 경우와 같음

# **apply lambda 식으로 데이터 가공**
판다스는 apply 함수에 lambda 식을 결합해 DataFrame이나 Series의 레코드 별로 데이터를 가공하는 기능을 제공한다.

```python
def get_square(a):
  return a**2

print('3의 제곱은:', get_square(3))

lambda_square= lambda x : x**2
print('3의 제곱은:', lambda_square(3))
```

* lambda 식을 이용할 때 여러 개의 값을 입력 인자로 사용해야 할 경우에는 보통 map() 함수를 결합해 사용

```python
a=[1,2,3]
squares = map(lambda x : x**2, a)
list(squares)
```

```python
titanic_df['Name_len']=titanic_df['Name'].apply(lambda x: len(x))
titanic_df[['Name_len', 'Name']].head(3)
```

```python
# 나이가 15세 미만이면 'Child', 그렇지 않으면 'Adult'로 구분하는 새로운 칼럼 
titanic_df['Child_Adult']=titanic_df['Age'].apply(lambda x : 'child' if x<=15 else 'Adult')
titanic_df[['Child_Adult', 'Age']].head(8)
```

<주의해야할 점>
* if 절의 경우 if 식보다 반환 값을 먼저 기술해야한다. 이는lambda 식 ':' 기호의 오른 편에 반환 값이 있어야하기 때문
* if, else만 지원하고, if, else if, else는 지원하지 않는다. else if를 사용하고 싶다면 else 절을 ()로 내포해 다시 if else()를 적용한다.

```python
titanic_df['Age_cat']=titanic_df['Age'].apply(lambda x : 'Child' if x<=15 else ('Adult' if x<=60 else 'Elderly'))
titanic_df[['Age','Age_cat']].head(8)
titanic_df['Age_cat'].value_counts()
```

```python
# 나이에 따라 세분화된 분류를 수행하는 함수 생성하기 
def get_category(age):
  cat=''
  if age<=5 : cat='Baby'
  elif age<=12: cat='Child'
  elif age<=18: cat='Teenager'
  elif age<=25: cat='student'
  elif age<=35: cat='Young Adult'
  elif age<=60: cat='Adult'
  else: cat='Elderly'
  return cat

titanic_df['Age_cat']=titanic_df['Age'].apply(lambda x :get_category(x))
titanic_df[['Age','Age_cat']].head()
```
